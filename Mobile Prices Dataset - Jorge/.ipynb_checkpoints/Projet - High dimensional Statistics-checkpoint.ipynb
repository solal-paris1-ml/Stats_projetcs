# <span style="color:green"> Etape 3 : Analyse Bivariée <span>

Nous pouvons à présent procéder à une analyse bivariée de nos données. Cette étape est essentielle, elle permet de repérer des liaisons entre les variables. Or, nous savons que des corrélations fortes entre des variables peuvent biaiser les résultats de certains modèles (notamment de la régression logsitique), d'où l'importance de cette étape. 

## 3.1. Etude de liaison entre les variables quantitatives.

Dans cette partie du document, nous étudions la liaison entre les variables quantitatives. 
Pour ce faire, nous utiliserons comme indicateurs la matrice des coefficients de corrélation, ensuite nous réaliserons le graphique du nuage de points pour les couples de variables, afin de visualiser les liaisons qui ne seraient pas linéaires entre des couples des variables, et enfin, pour confirmer ces potentielles liaisons nous réaliserons des tests statistiques.

### 3.1.1. Matrice des coefficients de correlations.

```python
#Matrice de corrélation
col = [ 1, 2, 4, 5, 6, 7, 12, 13]
plt.figure(figsize=(8,8))
masque  =  np.tril(data.iloc[: , col].corr())
sns.heatmap(data.iloc[: , col].corr(),annot=True,annot_kws={"size": 9}, vmin=-1, vmax=1,fmt='.2f',cmap= 'coolwarm' ,square=True,mask = masque)
plt.title('Matrice de corrélation des variable quantitative \n')
plt.show()
```

On constate donc qu'il n'y a aucune corrélation entre les variables quantitatives. 

### 3.1.2. Graphique pour visualiser la liaison entre les variables quantitatives :

```python
sns.pairplot(data.iloc[: , col]);
```

### 3.1.3. Tests statitisques pour confirmer les liaisons

Pour confirmer les liaisons existant entre les variables ci-dessous, nous réalisons le test du coefficient de corrélation linéaire de Pearson à 0.

```python
# Tests statistiques pour confirmer la liaison entre les variables
# Matrice des p-values (pearson)
data.iloc[:,col].pairwise_corr()
```

Le coefficient de corrélation de Pearson mesure la relation linéaire entre deux ensembles de données. À proprement parler, la corrélation de Pearson nécessite que chaque ensemble de données soit normalement distribué. Les corrélations de -1 ou +1 impliquent une relation linéaire négative et positive parfaite, respectivement, 0 indiquant l'absence d'association.
Dans notre cas on voit qu'il n'existe aucune relation linéaire entre les variables quantitatives.
